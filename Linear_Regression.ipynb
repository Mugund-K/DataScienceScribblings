```
from sklearn.datasets import make_regression
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```
X,y = make_regression(n_samples=100, n_features=1, noise=30)
data = pd.DataFrame({'X': X[:, 0], 'y': y})
sns.regplot(x='X', y='y', data=data, fit_reg=False)
```

```
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression()
linear_regressor.fit(X,y)
predicted = linear_regressor.predict(X)
print(predicted)
```

```
predicted_data= pd.DataFrame({'X': X[:, 0], 'predicted': predicted})
```

```
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
sns.regplot(x='X', y='y', data=data, fit_reg=False, ax=ax[0], color='green')
sns.lineplot(x='X', y='predicted', data=predicted_data, ax=ax[0], color='red')
sns.regplot(x='X', y='y', data=data, fit_reg= True, ax= ax[1], color='green')
```
